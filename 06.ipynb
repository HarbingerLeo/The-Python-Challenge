channel/readme.txt:

channel/90052.txt:

```python
visited = ["90052.txt"]
with open("Misc/channel/90052.txt") as f:
    content = [line.split(' ') for line in f.readlines()]
print(content)
while len(content) == 1 and content[0][0:3] == ['Next', 'nothing', 'is']:
    nextnum = content[0][3] + ".txt"
    path = "Misc/channel/{}".format(nextnum)
    with open(path) as f:
        content = [line.split(' ') for line in f.readlines()]
    visited.append(nextnum)
    print(content)
```

```python
# 记录遍历顺序，可能会有用？
print(visited)
```

最终遍历到`46145.txt`时提示收集注释，但是txt哪来的注释？

网页标签上的`now there are pairs`目前还没明白意思。

提示说答案在zip文件内，而不是解压后的文件夹？

考虑对zip文件本身做操作，python有一些处理zip文件的库，比如zlib，zipfile。

亦或者使用`zip()`函数？这样也对的上`pairs`这个提示

```python
from zipfile import ZipFile, ZipInfo

comments = {}
with ZipFile("Misc/channel.zip") as fzip:
    for file in fzip.infolist():
        comments[file.filename] = file.comment
print(comments)
```

可以看到，虽然不知道手法，但是每个txt文件确实都有一位注释！

接下来将尝试按上面的遍历顺序串起注释，这时`pairs`和`zip`的意味就很明显了。

```python
hidden = [comments[i] for i in visited]
bhid = b''.join(hidden)
print(bhid.decode())
```
