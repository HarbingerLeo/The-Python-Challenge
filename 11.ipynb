```python
from IPython.display import display
from PIL import Image
import requests
```

```python
with Image.open(
        requests.get(
            "http://huge:file@www.pythonchallenge.com/pc/return/cave.jpg",
            stream=True).raw) as im:
    print(im.size, im.mode)
    display(im)
```

题图如上，标签提示`odd even`，可以看出这图确实是亮的像素与暗的像素交叉在一起了，需要按奇偶拆开。

原图宽640高480，所以拆出的两个图应该都是宽320高240。

```python
# 第一个点是黑色，归给偶图
# 计算像素的坐标 x+y ，和为奇就归给奇，反之给偶
odd = Image.new(mode="RGB", size=(320, 240))
even = Image.new(mode="RGB", size=(320, 240))
for x in range(640):
    for y in range(480):
        pixel = im.getpixel((x, y))
        if (x + y) % 2 == 1:
            # to even
            even.putpixel((x // 2, y // 2), pixel)
        else:
            # to odd
            odd.putpixel((x // 2, y // 2), pixel)
display(odd, even)
```
