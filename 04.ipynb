```python
import requests

para = {"nothing": "8880"}
returntext = None
while returntext == None or (len(returntext) ==6 and returntext[0] == "and"):
    response = requests.get(
        "http://www.pythonchallenge.com/pc/def/linkedlist.php?", params=para)
    print(response.text)
    returntext = response.text.split(' ')
    para["nothing"] = returntext[-1]
```
