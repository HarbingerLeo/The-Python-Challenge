````python
import pickle
import requests

connect = requests.get("http://www.pythonchallenge.com/pc/def/banner.p")
data = connect.content
plain = pickle.loads(data)
print(plain)
````

````python
for ele in plain:
    row = ""
    for tup in ele:
        row += tup[0] * tup[1]
    print(row)
````
